```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle


# Read the CSV file into a DataFrame
df = pd.read_csv('output_clusters.csv')

# Separate points based on clusterId
clusters = df['clusterId'].unique()

print("Unique Cluster Labels:", df['clusterId'].unique())

# Plot each cluster with a different color
for cluster in clusters:
    cluster_points = df[df['clusterId'] == cluster]
    plt.scatter(cluster_points['x'], cluster_points['y'], s=10)
    
    plt.gcf().set_size_inches(10, 6)  # Specify width and height in inches


plt.title('Clustered Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()


plt.show()
```
